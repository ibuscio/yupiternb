import bs4
import requests

response = requests.get("https://www.platzi.com/cursos")
soup = bs4.BeautifulSoup(response.text, "html.parser") 

courses_link = soup.select(".CourseCategories-item")
courses = [course["href"] for course in courses_link]

for course in courses:
    print(course)
